```python
#Creating Categories for Rankings

rating_score = []
for score in df1['rating']:
    if score < 2.0 :
      rating_score.append('Poor')
    elif score < 3.0 :
      rating_score.append('Below Average')
    elif score < 4.0 :
      rating_score.append('Average')
    elif score < 5.0 :
       rating_score.append('Above Average')
    elif score == 5.0 :
       rating_score.append('Excellent')
```

Created a a Rating Category that consists of:

1. **Score below 2.0 = Poor**

2. **Score range of 2.0 - 2.9 = Below Average**

3. **Score range of 3.0 - 3.9 = Average**

4. **Score Range of 4.0 - 4.9 = Above Average**

5. **Score of 5.0 = Excellent**

```python
#Creating A new Column and Changing the Data Type
df1['rating_score'] = rating_score
df1['rating_score'] = df1['rating_score'].astype('category')
```

```python
# Define the desired order of categories
category_order = ['Below Average', 'Average', 'Above Average', 'Excellent']
# Reorder the categories using Categorical constructor
df1['rating_score'] = pd.Categorical(df1['rating_score'], categories=category_order, ordered=True)
```

1)Discounted Price and Other Variables:

The discounted price has a strong positive correlation with the difference in price (0.76). This suggests that as the discounted price increases, the price difference also tends to increase.
There is a moderate negative correlation between discounted price and discount percentage (-0.24). This indicates that higher discounts are associated with lower discounted prices.
The correlation with the rating is positive but relatively weak (0.12), suggesting a slight positive relationship between the discounted price and the product rating.

2) Discount Percentage and Other Variables:

The discount percentage has a negative correlation with both the discounted price (-0.24) and the product rating (-0.16). This implies that higher discount percentages are associated with lower discounted prices and lower ratings.

3) Rating and Other Variables:

The product rating shows a weak positive correlation with both the discounted price (0.12) and the rating count (0.10). This suggests that there is a slight tendency for products with higher discounted prices and more ratings to have higher ratings.

4) Rating Count and Other Variables:

The rating count has a weak negative correlation with the difference in price (-0.04). This indicates a slight tendency for products with higher price differences to have slightly fewer ratings.

5) Difference in Price and Other Variables:

The difference in price has a positive correlation with both the discounted price (0.76) and the discount percentage (0.09). This implies that higher price differences are associated with higher discounted prices and, to a lesser extent, higher discount percentages.
